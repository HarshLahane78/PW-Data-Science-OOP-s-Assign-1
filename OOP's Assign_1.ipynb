```python
class BankAccount :
    
    def __init__(self,Account_no,Account_holder,email_id,Deposit_amount,Withdraw_amount,Balance=0.0):
        
        self.Account_no = Account_no
        self.Account_holder = Account_holder
        self.email_id = email_id
        self.Deposit_amount = Deposit_amount
        self.Withdraw_amount = Withdraw_amount
        self.Balance =  Balance
        
    def OpenAccount(self):
        print("We are here at the bank to open a bank account")
        return self.Account_no,self.Account_holder,self.email_id,"\n"
    
    def Deposit(self,amount):
        
        print("\nWe are trying to deposit amount in the bank account")
        
        if amount > 0 :
            self.Balance += amount
            
            print(f"Depositing Amount : {amount} , Update Balance : {self.Balance}\n ")
            
        else :
            print("The entered amount is invalid . please enter the valid amount")
            
    
    def Withdraw(self,amount):
        
        print("We are trying to withdraw amount from the bank account")
        
        if amount <= self.Balance :
            self.Balance -= amount
            print(f"Withdrawing Amount : {amount} , Update Balance : {self.Balance} \n")
            
        else :
            print("Not Enough Balance inyour account")
            
        
    def Account_Details(self):
        print("Here are all the details available about the bank account")
        return self.Account_no,self.Account_holder,self.email_id,self.Balance,"\n"
    
    def get_balance(self):
        return f"Your Current Balance : {self.Balance}\n"
    
account1 = BankAccount(1234,"Harsh","harsh1@yahoo.com",10000,5000)

account1.OpenAccount()
account1.Deposit(10000)   
account1.Withdraw(5000)
account1.Account_Details()        
    
```

```python
class VehicleRentalSystem:
    
    def __init__(self, rent_date, return_date, available_vehicles, vehicle=None, rent_amount=0):
        self.available_vehicles = set(available_vehicles)
        self.rent_amount = rent_amount
        self.vehicle = vehicle
        self.rent_date = rent_date
        self.return_date = return_date
        
    def is_available(self):
        if self.vehicle in self.available_vehicles:
            print(f"{self.vehicle} is available\n")
        else:
            print("Vehicle is not available\n")
        
    def rent(self):
        if self.vehicle in ["bike1","bike2"]:
            self.rent_amount = 5000
            print(f"The rent of the vehicle will be: {self.rent_amount}\n")
        elif self.vehicle in ["car1","car2"]:
            self.rent_amount = 10000
            print(f"The rent of the vehicle will be: {self.rent_amount}\n")
        else:
            print("Invalid vehicle. Please enter the valid vehicle \n")
            
    def rent_vehicle(self):
        if self.vehicle in self.available_vehicles:
            self.available_vehicles.remove(self.vehicle)
            print(f"{self.vehicle} rented successfully\n")
        else:
            print("Vehicle is not available\n")
            
    def return_vehicle(self):
        self.available_vehicles.add(self.vehicle)
        print(f"{self.vehicle} returned successfully\n")
        
    def vehicle_details(self):
        print("details of the vehicle : ")
        print(f"Vehicle: {self.vehicle}")
        print(f"Rent of the vehicle: {self.rent_amount}\n")
        
    def get_available_vehicles(self):
        return self.available_vehicles
        

rent_sys1 = VehicleRentalSystem("24-12-2023", "1-1-2024", ["car1", "car2", "bike1", "bike2"])

rent_sys1.vehicle = "bike1"
rent_sys1.is_available()
rent_sys1.rent()
rent_sys1.rent_vehicle()
print(rent_sys1.get_available_vehicles())
rent_sys1.return_vehicle()
rent_sys1.vehicle_details()
```

```python
class StudentManagement :
    
    def __init__(self,student_id,student_name,student_grade):
        
        self.student_id = student_id
        self.student_name = student_name
        self.student_grade = student_grade
        
    def grade_calculation(self) :
        
        print("Student Name : ",self.student_name)
    
        if 100 >= self.student_grade >= 90 :
            print("Student Grade : Grade A+")
            
        elif 89 >= self.student_grade >= 80 :
            print("Student Grade : Grade A")
            
        elif 79 >= self.student_grade >= 70 :
            print("Student Grade : Grade B")
            
        elif 69 >= self.student_grade >= 60 :
            print("Student Grade : Grade C")
            
        elif 59 >= self.student_grade >= 35 :
            print("Student Grade : Grade D")
            
        elif 0 > self.student_grade :
            print("Invalid grade . Please enter the valid grades")
            
        else :
            print("Invalid grade . Please enter the valid grades")
            
            
    def student_details(self):
        print("\nStudent details : \n")
        print(f"Student Name : {self.student_name}")
        print(f"Student ID : {self.student_id}")
        print(f"Student Grade : {self.student_grade}")
        

student1 =  StudentManagement(1234,"HARSH",67)
student1.grade_calculation()
student1.student_details()
            
```

```python
class SocialMediaProfile :
    
    def __init__(self,username):
        self.username = username
        self.posts = []
        
    def UploadPost(self,content):
        self.posts.append(content)
        print("Post uploaded successfully")
        
    def DisplayPost(self):
        
        if not self.posts :
            print(f"\nPosts are not available by username : {self.username}")
            
        else :
            print(f"\nAvailable posts on the profile by username : {self.username}")
            for i,post in enumerate(self.posts,start=1):
                print(f"{i} : {post}")
                
                
    def SearchPost(self,keyword):
        
        matching_posts = (post for post in self.posts if keyword.lower() in post.lower())
        
        if not matching_posts :
            print(f"\nThere are no matching post available with the keyword : '{keyword}' in the username : {self.username}")
            
        else :
            print(f"\nMatching posts for keyword : '{keyword}' in the username : {self.username} ")
            for i,post in enumerate(matching_posts,start=1):
                print(f"{i} : {post}")
                
user_profile1 =  SocialMediaProfile("harsh_123")


user_profile1.UploadPost("We are finally graduated....")
user_profile1.UploadPost("Now I need to find a job....")
user_profile1.UploadPost("live in peace....")     

user_profile1.DisplayPost()

user_profile1.SearchPost("peace")

```

```python
class ToDoList :
    
    def __init__(self):
        self.tasks = {}
        
    def AddTask(self,task,due_date):
        print("Adding Tasks : ")
        self.tasks[task] = due_date
        print(f"Task : {task} added successfully in the ToDo List with due date : {due_date}\n")
        
    def mark_as_completed(self,task):
        
        if task in self.tasks :
            del self.tasks[task]
            print(f"Task : {task} is marks as completed\n ")
            
        else : 
            print("Task is not available in the ToDo List\n")
            
    def display_pending_tasks(self):
        
        if not self.tasks :
            print("No Pending Tasks")
            
        else :
            print("Pending Tasks : \n")
            
            for task , due_date in self.tasks.items():
                print(f"Task : {task} , Due Date : {due_date}")
                

todo_list = ToDoList()

todo_list.AddTask("Complete project", "2024-01-15")
todo_list.AddTask("Buy groceries", "2024-01-20")
todo_list.AddTask("Read a book", "2024-01-25\n")

todo_list.display_pending_tasks()

todo_list.mark_as_completed("Buy groceries")

todo_list.display_pending_tasks()
```
